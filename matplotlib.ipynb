### Matplotlib workflow

```python
#1. Setup imports
%matplotlib inline
import matplotlib.pyplot as plt

#2.Setup data
x=[1,2,3,4,5]
y=[50,100,150,200,250]

#3. Setup plot
fig, ax= plt.subplots(figsize=(10,10))
ax.set(title="Example plot",
       xlabel="x-axis",
       ylabel="y-axis")
#4. Plot data
ax.plot(x,y)

#5. Save plot
fig.savefig("plot.png")
```

## Different types of plots

* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
x=np.linspace(1,10,100, dtype=int)
x
```

```python
#1. Line plot
fig, ax=plt.subplots()
ax.plot(x,x**2)
ax.set(title="Line plot",
       xlabel="x values",
       ylabel="square of x");
```

```python
fig, ax=plt.subplots()
ax.plot(x,np.exp(x))
ax.set(title="Line plot",
       xlabel="x values",
       ylabel="exp of x");
```

```python
fig, ax=plt.subplots()
ax.plot(x,np.log(x))
ax.set(title="Line plot",
       xlabel="x values",
       ylabel="log of x");
```

```python
y=np.linspace(0,10,100)
fig, ax=plt.subplots()
ax.plot(y,np.sin(y))
ax.set(title="Line plot",
       xlabel="y values",
       ylabel="sine of y");
```

```python
#2. Scatter plot
y=np.linspace(0,10,100)
fig, ax=plt.subplots()
ax.scatter(y,np.sin(y))
ax.set(title="Line plot",
       xlabel="y values",
       ylabel="sine of y");
```

```python
#3.Bar plot
butter_prices={"Almond":10,
               "Peanut":8,
               "Cashew":12}
fig,ax=plt.subplots()
ax.bar(butter_prices.keys(),butter_prices.values())
ax.set(title="Bar graph",
       xlabel="Butter names",
       ylabel="Prices ($)");
```

```python
#Horizontal bar
butter_prices={"Almond":10,
               "Peanut":8,
               "Cashew":12}
fig,ax=plt.subplots()
ax.barh(list(butter_prices.keys()),list(butter_prices.values()))
ax.set(title="Bar graph",
       xlabel="Butter names",
       ylabel="Prices ($)");
```

```python
#4 Histogram
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

```python
#5 Subplots
# This involves multiple plots
```

```python
#Way 1
fig,((ax0,ax1),(ax2,ax3))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))
ax0.plot(np.linspace(0,10,100))
ax1.bar(butter_prices.keys(),butter_prices.values())
ax2.hist(np.random.randn(100))
ax3.scatter(np.random.random(10),np.random.random(10))
```

```python
#Way 2
fig,ax=plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(10,5))
ax[0,0].plot(np.linspace(0,10,100))
ax[0,1].bar(butter_prices.keys(),butter_prices.values())
ax[1,0].hist(np.random.randn(100))
ax[1,1].scatter(np.random.random(10),np.random.random(10))
```

```python
car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Sale Date"]=pd.date_range('1/1/2020',periods=10)
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','')
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales.Price=car_sales.Price.astype(int)
car_sales.dtypes
```

```python
car_sales
```

```python
car_sales["Total Price"]=car_sales.Price.cumsum()
car_sales
```

```python
fig,ax=plt.subplots()
ax.plot(car_sales["Sale Date"],car_sales["Total Price"])
```

```python
car_sales.plot(x="Sale Date",y="Total Price");
```

## Heart disease dataset

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease.age.plot(kind="hist");
```

```python
heart_disease.plot(kind="scatter",
                   x="age",
                   y="chol",
                   c="target")
```

```python
fig , ax = plt.subplots(figsize=(10,6))
heart_disease.plot(kind="scatter",
                   x="age",
                   y="chol",
                   c="target",
                   ax=ax)
ax.set_xlim([45,100])
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
len(over_50)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter=ax.scatter(over_50["age"],
                   over_50["chol"],
                   c=over_50["target"])
ax.set(title="Cholestrol over age",
       xlabel="age",
       ylabel="chol")
ax.legend(*scatter.legend_elements(),title="Target")
ax.axhline(over_50.chol.mean(),linestyle='--')
```

```python
over_50.head()
```

```python
fig, (ax0,ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                           sharex=True)
scatter0=ax0.scatter(x=over_50.age,
                     y=over_50.chol,
                     c=over_50.target)
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(over_50.chol.mean(),linestyle='--')
ax0.set(title="Heart Disease and Cholestrol",
       ylabel="Chol")
scatter1=ax1.scatter(x=over_50.age,
                     y=over_50.thalach,
                     c=over_50.target)
ax1.axhline(over_50.thalach.mean(),linestyle='--')
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.set(title="Heart Disease and Max Heart Rate",
        ylabel="Heart Rate",
        xlabel="Age")
fig.suptitle("Heart Disease Classification",fontsize=16,fontweight="bold")
```
